## Ahora hablemos de Estructura

Tomado de [Astera](https://www.astera.com/es/type/blog/structured-semi-structured-and-unstructured-data/)

1.  **Datos Estructurados**: Los datos estructurados son informacin que ha sido formateada y transformada en un modelo de datos bien definido. Los datos sin procesar se mapean en campos prediseados que luego se pueden extraer y leer a travs de SQL fcilmente. Las bases de datos relacionales SQL, que consisten en tablas con filas y columnas, son el ejemplo perfecto de datos estructurados.

2. **Datos semiestructurados**:  o los datos parcialmente estructurados son otra categora entre los datos estructurados y no estructurados. Los datos semiestructurados son un tipo de datos que tienen algunas caractersticas consistentes y definidas. No se limita a una estructura rgida como la necesaria para las bases de datos relacionales. Las propiedades organizativas como los metadatos o las etiquetas semnticas se utilizan con datos semiestructurados para hacerlos ms manejables; sin embargo, todava contiene cierta variabilidad e inconsistencia.

3. **No estructurado**: Los datos se definen como datos presentes en forma absoluta sin procesar. Estos datos son difciles de procesar debido a su compleja organizacin y formato. La gestin de datos no estructurados puede tomar datos de muchas formas, incluidas publicaciones en redes sociales, chats, imgenes satelitales, datos de sensores de IoT, correos electrnicos y presentaciones, para organizarlos de manera lgica y predefinida en un almacenamiento de datos.
